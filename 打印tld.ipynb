```python
def load_data(filename):
    if 'txt' in filename or 'log' in filename or 'json' in  filename:
        ip_list=[]
        with open(filename,'r') as f:
            ip_list=f.readlines()
        return [i.strip() for i in ip_list]
    else:
        import pickle
        with open(filename,'rb') as f:
            return pickle.load(f)
```
